```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Load the data from the CSV file
data = pd.read_csv('/home/manichandana-sandhaboina/Downloads/archive(4)/CompanyABCProfit.csv')

# Assuming the data is in a column named 'Profit(Rs \'000)'
values = data['Profit(Rs \'000)']

# Calculate the mean and standard deviation of the data
mean = values.mean()
std = values.std()

# Standardize the data (z-scores)
standardized_values = (values - mean) / std

# Calculate standardized mean and standard deviation
standardized_mean = standardized_values.mean()
standardized_std = standardized_values.std()

# Plot histogram of original data
plt.hist(values, bins=30, density=True, alpha=0.5, label='Data')

# Plot standard normal distribution (bell curve)
x = np.linspace(mean - 3*std, mean + 3*std, 100)
plt.plot(x, norm.pdf(x, mean, std), 'r', lw=2, label='Standard Normal')

plt.title('Standard Normal Distribution and Data Histogram')
plt.xlabel('Values')
plt.ylabel('Density')
plt.legend()
plt.show()

# Print mean and standard deviation
print(f"Original Mean: {mean}")
print(f"Original Std Dev: {std}")

# Print standardized mean and standard deviation
print(f"Standardized Mean: {standardized_mean:.6f}")
print(f"Standardized Std Dev: {standardized_std:.6f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-10, 10, 100)  # from -10 to 10, with 100 points

# Compute sigmoid and its derivative
sigmoid = 1 / (1 + np.exp(-x))
sigmoid_derivative = sigmoid * (1 - sigmoid)

# Plotting
plt.figure(figsize=(10, 6))

# Sigmoid function plot
plt.subplot(2, 1, 1)
plt.plot(x, sigmoid, label='Sigmoid')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')

plt.legend()

# Derivative of sigmoid function plot
plt.subplot(2, 1, 2)
plt.plot(x, sigmoid_derivative, label='Sigmoid Derivative', color='orange')
plt.title('Derivative of Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid\'(x)')

plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-10, 10, 200)

# Compute tanh and its derivative
tanh_vals = np.tanh(x)
tanh_deriv_vals = 1 - np.tanh(x)**2

# Plotting
plt.figure(figsize=(10, 6))

# Tanh function plot
plt.subplot(2, 1, 1)
plt.plot(x, tanh_vals, label='tanh(x)')
plt.title('Tanh Function')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.grid(True)
plt.legend()

# Tanh derivative plot
plt.subplot(2, 1, 2)
plt.plot(x, tanh_deriv_vals, label='tanh\'(x)', color='orange')
plt.title('Derivative of Tanh Function')
plt.xlabel('x')
plt.ylabel('tanh\'(x)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-10, 10, 200)

# Compute ReLU and its derivative
relu_vals = np.maximum(0, x)
relu_deriv_vals = np.where(x > 0, 1, 0)

# Plotting
plt.figure(figsize=(10, 6))

# ReLU function plot
plt.subplot(2, 1, 1)
plt.plot(x, relu_vals, label='ReLU')
plt.title('ReLU Function')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.legend()

# Derivative of ReLU function plot
plt.subplot(2, 1, 2)
plt.plot(x, relu_deriv_vals, label='ReLU\'(x)', color='orange')
plt.title('Derivative of ReLU Function')
plt.xlabel('x')
plt.ylabel('ReLU\'(x)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Leaky ReLU function and its derivative parameters
alpha = 0.01
# Generate x values
x = np.linspace(-10, 10, 200)
# Compute Leaky ReLU and its derivative
leaky_relu_vals = np.where(x > 0, x, alpha * x)
leaky_relu_deriv_vals = np.where(x > 0, 1, alpha)
# Plotting
plt.figure(figsize=(10, 6))
# Leaky ReLU function plot
plt.subplot(2, 1, 1)
plt.plot(x, leaky_relu_vals, label='Leaky ReLU')
plt.title('Leaky ReLU Function')
plt.xlabel('x')
plt.ylabel('Leaky ReLU(x)')
plt.legend()
# Derivative of Leaky ReLU function plot
plt.subplot(2, 1, 2)
plt.plot(x, leaky_relu_deriv_vals, label='Leaky ReLU\'(x)', color='orange')
plt.title('Derivative of Leaky ReLU Function')
plt.xlabel('x')
plt.ylabel('Leaky ReLU\'(x)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define softmax function
def softmax(x):
    e_x = np.exp(x - np.max(x))  # Subtracting max(x) for numerical stability
    return e_x / e_x.sum()

# Define softmax derivative function (numerically)
def softmax_derivative(x):
    s = softmax(x)
    return s * (1 - s)

# Generate data points
x = np.linspace(-5, 5, 100)
softmax_values = softmax(x)
softmax_der = softmax_derivative(x)

# Plotting both softmax function and its derivative
plt.figure(figsize=(14, 6))

# Plot softmax function
plt.subplot(1, 2, 1)
plt.plot(x, softmax_values, label='Softmax', color='blue')
plt.title('Softmax Activation Function')
plt.xlabel('Input')
plt.ylabel('Output Probability')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()

# Plot softmax derivative
plt.subplot(1, 2, 2)
plt.plot(x, softmax_der, label='Softmax Derivative', color='red')
plt.title('Derivative of Softmax Activation Function')
plt.xlabel('Input')
plt.ylabel('Derivative')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()

# Adjust layout to prevent overlap
plt.tight_layout()

# Show plot
plt.show()
```
